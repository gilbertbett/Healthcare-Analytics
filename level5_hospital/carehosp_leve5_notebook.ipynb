```python
import pandas as pd
```

```python
# Load datasets into pandas DataFrames

patients_df = pd.read_csv('data/patients.csv')
visits_encounters_df = pd.read_csv('data/visits_encounters.csv')
labs_orders_df = pd.read_csv('data/labs_orders.csv')
labs_results_df = pd.read_csv('data/labs_results.csv')
diagnoses_df = pd.read_csv('data/diagnoses.csv')
clinical_outcomes_df = pd.read_csv('data/clinical_outcomes.csv')
adt_events_df = pd.read_csv('data/adt_events.csv')
billing_lines_df = pd.read_csv('data/billing_lines.csv')
tarrifs_services_df = pd.read_csv('data/tariffs_services.csv')
```
